```python
#Plotting graph 
#STOCK NAME - IRCTC
#PERIOD - 3/01/2022 TO 30/12/2022
#TIME FRAME - 1 DAY
import plotly.graph_objects as go

fig = go.Figure(data=go.Scatter(x=data.index,y=data['Close'], mode='lines'))
fig.update_layout(title={'text':'IRCTC 1DAY TIMEFRAME', 'x':0.5})
fig.show()
```

```python
from ipywidgets.widgets.widget_color import Color
#Adding trading volumes in previous chart
#STOCK NAME - IRCTC
#PERIOD - 3/01/2022 TO 30/12/2022
#TIME FRAME - 1 DAY
from plotly.subplots import make_subplots

# assign colors to type using a dictionary
colors = {'A':'steelblue',
          'B':'firebrick'}

fig2 = make_subplots(specs=[[{"secondary_y": True}]])
fig2.add_trace(go.Scatter(x=data.index,y=data['Close'],name='Price'),secondary_y=False)
fig2.add_trace(go.Bar(x=data.index,y=data['Volume'],name='Volume',marker_color=colors['B']),secondary_y=True, )
fig2.update_layout(title={'text':'IRCTC 1DAY TIMEFRAME', 'x':0.5})
fig2.show()
```


We will modify the volume a little bit. Right now the volume bars all have the same color(total volume). We can use different colors to distinguish between an up or down day – green for up days, and red for down days.

To do that, we just need to calculate the daily change (positive or negative) then insert a color column into our dataframe. Then we can pass the color information into the volume data series. The marker argument dictates how our scatter plot should look like – color, shape, size, etc

```python
#Chandelstick chart
fig3 = make_subplots(specs=[[{"secondary_y": True}]])
fig3.add_trace(go.Candlestick(x=data.index,
                              open=data['Open'],
                              high=data['High'],
                              low=data['Low'],
                              close=data['Close'],
                             ))
fig3.update_layout(title={'text':'IRCTC 1DAY TIMEFRAME', 'x':0.5})
```

#Indicators

#We’ll draw a simple indicator 20 Day Moving Average here to show the concept, theoretically, we can plot any indicator on the chart.

The df.rolling() method provides “moving windows” that we can operate on. To get the average of the moving window, we just need to add the .mean() at the end of the rolling() method.

```python
fig3.add_trace(go.Scatter(x=data.index,y=data['Close'].rolling(window=20).mean(),marker_color='blue',name='20 Day MA'))
fig3.add_trace(go.Bar(x=data.index, y=data['Volume'], name='Volume'),secondary_y=True)
fig3.update_layout(title={'text':'IRCTC 1DAY TIMEFRAME', 'x':0.5})
fig3.update_yaxes(range=[0,1000000000],secondary_y=True)
fig3.update_yaxes(visible=False, secondary_y=True)
fig3.update_layout(xaxis_rangeslider_visible=False)  #hide range slider
fig3.show()
```

```python
#TAKING DIFFERENCE OF OPENING AND CLOSING PRICE 
#ALSO ADDING COLOR GREEN FOR POSITIVE RETURN IN A DAY, RED - FOR NEGATIVE RETURN FOR A DAY
data['diff'] = data['Close'] - data['Open']
data.loc[data['diff']>=0, 'color'] = 'green'
data.loc[data['diff']<0, 'color'] = 'red'
data.head()
```

```python
fig3 = make_subplots(specs=[[{"secondary_y": True}]])
fig3.add_trace(go.Candlestick(x=data.index,
                              open=data['Open'],
                              high=data['High'],
                              low=data['Low'],
                              close=data['Close'],
                              name='Price'))
fig3.add_trace(go.Scatter(x=data.index,y=data['Close'].rolling(window=20).mean(),marker_color='blue',name='20 Day MA'))
fig3.add_trace(go.Bar(x=data.index, y=data['Volume'], name='Volume', marker={'color':data['color']}),secondary_y=True)
fig3.update_yaxes(range=[0,700000000],secondary_y=True)
fig3.update_yaxes(visible=False, secondary_y=True)
fig3.update_layout(xaxis_rangeslider_visible=False)  #hide range slider
fig3.update_layout(title={'text':'IRCTC 1DAY TIMEFRAME', 'x':0.5})
fig3.show()
```

The first 19 days are essentially blank. We neeed 20 days data for 20 days moving averages.

#Hide Non-trading Days

#Stock markets close on weekends and holidays, so there’s no data for those periods.

Plotly charts have a rangebreaks attribute that we can use to hide certain time periods. 

All we need is to add another update_axes(rangebreaks=[…]) to the above code, just before the fig.show()


bounds = [‘sat’, ‘mon’] will hide Saturdays and Sundays

bounds = [16, 9.5] will hide between 4pm to 9:30am, which are market closed hours

values = [”2022-01-01”,"2022-12-25"] can hide individual days

```python
fig3 = make_subplots(specs=[[{"secondary_y": True}]])
fig3.add_trace(go.Candlestick(x=data.index,
                              open=data['Open'],
                              high=data['High'],
                              low=data['Low'],
                              close=data['Close'],
                              name='Price'))
fig3.add_trace(go.Scatter(x=data.index,y=data['Close'].rolling(window=20).mean(),marker_color='blue',name='20 Day MA'))
fig3.add_trace(go.Bar(x=data.index, y=data['Volume'], name='Volume', marker={'color':data['color']}),secondary_y=True)
fig3.update_yaxes(range=[0,700000000],secondary_y=True)
fig3.update_yaxes(visible=False, secondary_y=True)
fig3.update_layout(xaxis_rangeslider_visible=False)  #hide range slider
fig3.update_layout(title={'text':'IRCTC 1DAY TIMEFRAME', 'x':0.5})


fig3.update_xaxes(rangebreaks = [
                       dict(bounds=['sat','sun']), # hide weekends
                        # for hourly chart, hide non-trading hours (24hr format)
                       dict(values=["2022-01-01", "2022-12-25"]) #hide Xmas and New Year
                                ])
fig3.show()
```
